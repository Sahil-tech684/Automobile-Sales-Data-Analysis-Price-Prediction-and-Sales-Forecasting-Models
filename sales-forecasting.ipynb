# **Time Series Analysis**

## Introduction
Time series analysis is a statistical technique used to analyze patterns and trends in data points collected and recorded over time. It is widely employed in various fields such as finance, economics, weather forecasting, signal processing, and many others. Time series data consists of observations or measurements recorded at different time intervals.

## **Time Series Modeling**

```python
df_time = df.groupby('date').size().reset_index(name='sales')
```

```python
df_time.set_index('date',inplace=True)
df_time
```

```python
df_time.plot(figsize=(18,8))
```

We want to predict Monthly sales so we will transform our data.

```python
df_monthly = df_time.resample('M').sum()
df_monthly.plot(figsize=(12,8))
```

### **Plotting Exponential Weighted Moving Average (EWMA)**
EWMA is a technique used to smooth out noisy data by giving more weight to recent observations while still considering older observations. This can help in visualizing underlying trends and patterns in the data by reducing the impact of short-term fluctuations or outliers.

By plotting the EWMA alongside the original time series data, you can visually identify trends more clearly. EWMA tends to react more quickly to changes in the data compared to traditional moving averages, making it useful for detecting shifts in trends over time.

```python
alpha = 0.5

df_monthly['ewma'] = df_monthly['sales'].ewm(alpha=alpha, adjust=False).mean()

month_year_labels = df_monthly.index.strftime('%b %Y')

plt.figure(figsize=(18, 8))
plt.plot(month_year_labels, df_monthly['sales'], label='Monthly Sales')
plt.plot(month_year_labels, df_monthly['ewma'], label=f'EWMA (alpha={alpha})')
plt.title('Exponential Weighted Moving Average (EWMA) for Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right') 
plt.legend()
plt.tight_layout()
plt.show()
```

### **Dickey Fuller Test**
- The Dickey-Fuller test is a statistical test commonly used in econometrics and time series analysis to determine whether a unit root is present in a time series dataset. A unit root indicates that a time series is non-stationary, meaning that its statistical properties (such as mean and variance) change over time.

- The main point of performing the Dickey-Fuller test is to assess whether a given time series is stationary or non-stationary.

```python
from statsmodels.tsa.stattools import adfuller

sales_series = df_monthly['sales']
result = adfuller(sales_series)

test_statistic, p_value, _, _, _, _ = result

print(f'Test Statistic: {test_statistic}')
print(f'P-value: {p_value}')

if p_value <= 0.05:
    print("Reject the null hypothesis: The time series is stationary.")
else:
    print("Fail to reject the null hypothesis: The time series is non-stationary.")
```

Since our data is non-stationary we are going to use a method called `Differencing` to make our data stationary.

```python
df_monthly2 = df_monthly.copy()
df_monthly2['sales_diff'] = df_monthly2['sales'] - df_monthly2['sales'].shift(1)
```

```python
df_monthly2 = df_monthly2.dropna()
df_monthly2
```

Again doing dickey fuller test

```python
sales_series = df_monthly2['sales_diff']
result = adfuller(sales_series)

test_statistic, p_value, _, _, _, _ = result

print(f'Test Statistic: {test_statistic}')
print(f'P-value: {p_value}')

if p_value <= 0.05:
    print("Reject the null hypothesis: The time series is stationary.")
else:
    print("Fail to reject the null hypothesis: The time series is non-stationary.")
```

```python
df_monthly2['sales_diff'].plot()
```


### **SARIMAX**
- Since our data have some seaonality so we will use `SARIMAX(Seaonal Autoregressive Integrated Moving Average eXogenous)` model.
- To use this model we have to determine `P,D,Q` values.
- We did differencing with only one shift to make our data stationary so our `D=1`.
- We will use `ACF(Autocorrelation Function)` and `PACF(Partial Autocorrelation Function)` plots to detemine `P` & `Q` values.

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

Identification of MA model is best done with `ACF` plot.

```python
acf_plot = plot_acf(df_monthly2['sales_diff'])
```

Identification of AR model is best done with `PACF` plot.

```python
pacf_plot = plot_pacf(df_monthly2['sales_diff'])
```

##### **After evaluation both `ACF` and `PACF` plots we can conclude that:**
- `P=1`, from `PACF` plot.
- `Q=1`, from `ACF` plot.

## **Model Training and Evaluation**

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_monthly, test_size=0.2, shuffle=False)
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from math import sqrt

order = (1, 1, 1) 
```

- We determined `P,D,Q` values above.
- For `SARIMAX` model we also have to determine a seasonal component which specifies the periodicity of the seasonal component, which is the number of time steps in each seasonal period.

```python
rmse_diff_values = []
for i in range(2,13):    
    model = sm.tsa.statespace.SARIMAX(train['sales'], order=order, seasonal_order=(1,1,1,i))
    result = model.fit(disp=False)
    start = len(train)
    end = len(train) + len(test) - 1
    predictions = result.predict(start=start, end=end, dynamic=True, typ='levels')
    rmse = sqrt(mean_squared_error(test['sales'], predictions))
    rmse_diff_values.append(rmse)
plt.plot(range(2, 13), rmse_diff_values, marker='o')
plt.xlabel('i')
plt.ylabel('RMSE')
plt.grid(True)
plt.show()
```

- The given code calculates the Root Mean Squared Error (RMSE) for different seasonal periods 'i' in the range from 2 to 12. It utilizes the SARIMAX model from the statsmodels library to perform time series analysis. 

- Finally, the code plots a line graph where the x-axis represents the range of 'i' (2 to 12), and the y-axis represents the corresponding RMSE values. 

- From the plot, it's observed that the minimum `RMSE` occurs when 'i' is equal to `6`.

- So, value of `seasonal component` will be `6`.

Fitting model with optimal parameters.

```python
model = sm.tsa.statespace.SARIMAX(train['sales'], order=order, seasonal_order=(1,1,1,6))
result = model.fit(disp=False)
start = len(train)
end = len(train) + len(test) - 1
predictions = result.predict(start=start, end=end, dynamic=True, typ='levels')
rmse = sqrt(mean_squared_error(test['sales'], predictions))
rmse
```

- `RMSE` of `100` can be considered as reasonable.
